```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid',context='notebook')
#cols = ['date','time','weekday','startstation','endstation','vehicletype','velocity','numbers']
#cols = ['velocity','numbers']
#sns.pairplot(dfM05A2016R10E7[cols],size = 2.5);
#plt.ylim(0,150)
#plt.show()
```

```python
#2016年隨機100萬筆，流量 vs 車速
plt.plot(dfM05A2016R10E7.numbers,dfM05A2016R10E7.velocity,marker='o',lw=0,color = 'g')
plt.xlabel('numbers')
plt.ylabel('velocity')
plt.xlim(0,500)
plt.ylim(0,150)
plt.show()
```

```python
#2016年隨機100萬筆，weekday 流量 vs 車速
for i in range(7):
    weekday12016 = dfM05A2016R10E7[dfM05A2016R10E7['weekday']==i+1]
    plt.plot(weekday12016.numbers,weekday12016.velocity,marker='o',lw=0,color='b')
    plt.xlim(0,500)
    plt.ylim(0,150)
    plt.xlabel('numbers')
    plt.ylabel('velocity')
    if(i==6):
        break
    plt.subplots()
plt.show()
```
