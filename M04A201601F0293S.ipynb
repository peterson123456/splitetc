```python
import pandas as pd
import matplotlib.pyplot as plt
dfm04a201601f0293s = pd.read_csv('M04A201601F0293S.csv',sep=',', names = ['datetime','startstation','endstation','vehicletype','traveltime','numbers'])
dfm04a201601f0293s.head()
```

```python
#2016年01F0264S~01F0293S 所有車種 旅行時間 vs 車流量
plt.plot(dfm04a201601f0293s.numbers,dfm04a201601f0293s.traveltime,marker='o',lw=0,color = 'g')
plt.xlabel('numbers')
plt.ylabel('traveltime')
plt.xlim(0,500)
plt.ylim(0,3000)
plt.show()
```
